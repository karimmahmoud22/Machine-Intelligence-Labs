## I. Python is Dynamically typed & Strongly typed

Dynamically Typed means that the interpreter assigns the type at runtime based on the variable's value. 

```python
salad = [1, 2.05, 1.1e18, True, "Hello World", '"Hello World"', 1+2j, [1, 2, 3], (1,), (1, 2, 3), {1, 2, 3}, {"a": 1, "b": 2, "c": 3}, None]
for x in salad:      print('x =', x, '\t\t\t=> with type', type(x))
```

We can use del to delete a reference to an object. Garbage collection will take care of the object if there are no other references to it. This is mostly used to delete an element from a list or dictionary.

```python
x = 1
print('x =', x, '\t\t=> with type', type(x))

del x
print('x =', x, '\t\t=> with type', type(x))
```

Strongly Typed means that any changes in the type must be explicit (unlike Javascript)

```python
x = "This number is " + 12
```

An Explicit Conversion:

```python
x = "This number is " + str(12)
print(x)
x = int("13") + 12
print(x)
```

# II. Python Operators

#### 1. Boolean Operations

```python
x, y = True, False
print('x and y = ', x and y)
print('x or y = ', x or y)
print('not x = ', not x)                        ### ! can be only used before =
print(' x != y', x != y)         
```

#### 2. Bitwise Operations

Note that an integer is returned in each case, we use bin() to convert it to a binary string.

```python
x, y = 0b0101, 0b0011
print('x & y = ', bin(x & y))
print('x | y = ', bin(x | y))
print('~ x = ', bin(~ x))
print('x ^ y = ', bin(x ^ y))             ### XOR
print('x << 1', bin(x << 1))              ### shift left (multiply by 2)
print('x >> 1', bin(x >> 1))              ### shift right (divide by 2)
```

#### 3. Assignment
Normal assignment operations will simply point the new variable towards the existing object.

```python
x = [1,2,3,4,5]
y = x
y[1] = 'edit'
print('x = ', x)
print('y = ', y)
```

When dealing with mutable objects, we need to think if we want to make a copy or not.

```python
x = [1,2,3,4,5]
y = x.copy()            ### Shallow copy. Slicing also does shallow copy.
y[1] = 'edit'
print('x = ', x)
print('y = ', y)
```

 In other cases (usually nesting), a deep copy might be needed so no references are shared due to nesting.

#### 8. Passing arguments
In Python, arguments are passed by assignment. So think of assigning each function argument to the corresponding input in the call.

```python
def do_changes(x,l):
    x += 1                  # Here x was assigned to a new value. It's no longer the same x that was passed in.
    l.append('hello')       # L was not assigned to a new value. It's the same l that was passed in.

x, l = 0, []
do_changes(x,l)
print(x, l)

def do_changes(x,l):
    x += 1
    l = l + ['hello']

x, l = 0, []
do_changes(x,l)
print(x, l)
```

#### 9. Passing Functions is Okay

```python
def dbl(x):
    return x*2
def dummy(x):
    return x
def apply_and_sum(l, fn=dummy):
    return sum(fn(i) for i in l)

print(apply_and_sum([1,2,3], dbl))
```

#### 10. Lambda Functions (One-liners)
This is eqiuvalent to the cell above.

```python
dbl = lambda x: x*2

def apply_and_sum(l, fn = lambda x: x):
    return sum(fn(i) for i in l)

print(apply_and_sum([1,2,3], dbl))
```

We can also have no or multiple arguments.

# Exercise
- write a reduction function that multiplies all elements of a list (reduction means "applying function on each element of the array, in order, passing in the return value from the calculation on the preceding element")

```python
# Write a reduce function
def reduce(fn, l):
    if len(l) == 1:
        return l[0]
    else:
        return fn(l[0], reduce(fn, l[1:]))

print(reduce(lambda x,y:x*y, [1,2,3,4]))      
   
```

The following is an example where x exists three times, each time in a different scope 

```python
def hello():
    x = 2
    
    def bye():
        x = 3
        print("inneeer", x)
    
    print("inner", x)
    bye()
    print("inner", x)
    

x=1
print("outer", x)
hello()
print("outer", x)
```

Make sure while nesting that all functions agree the variable is global

```python
def hello():                            # This doesn't (unless we specify global x)
    x = 2
    def bye():                          # This and the outer scope agree its global
        global x
        x = 3
        print("inneeer", x)
    
    print("inner before call", x)
    bye()
    print("inner after call", x)
    

x = 1
print("outer before call", x)
hello()
print("outer after call", x)
```

#### 4. Use nonlocal to share scope with the enclosing function

```python
def hello():
    x = 2
    
    def bye():
        nonlocal x
        x = 3
        print("inneeer", x)
    
    print("inner before call", x)
    bye()
    print("inner after call", x)
    

x=1
print("outer before call", x)
hello()
print("outer after call", x)
```

# XV. Common Mistakes

#### 1. Default Objects

```python
def addSeq(n, l=[]):
    while n != 0:
        l.append(n)
        n //= 2
    return l

print(addSeq(8, ['Hello']))         # Here we assigned l = ['Hello'] 
print(addSeq(8))                    # Here we assigned l = []
print(addSeq(8))                    # We didn't change the assignment of l, it still points to the same list
print(addSeq(8))                    # Also here
print(addSeq(8,[]))                 # Has nothing to do with the previous call, we assigned a new value
print(addSeq(8))                    # Goes back to the default
```

Solution: reassign the default everytime.

```python
def addSeq(n, l=None):
    l = l or []             
    while n != 0:
        l.append(n)
        n //= 2
    return l

print(addSeq(8, ['Hello']))
print(addSeq(8))
print(addSeq(8))
print(addSeq(8))
print(addSeq(8,[]))
print(addSeq(8))
```

#### 2. Multiplying Lists

```python
l1 = [5] * 4        # Here it will repeat the value 4 times
l1[0] = 7
print(l1)

l2 = [[]]*4       # When it repeats [] 4 times, they all have the same reference (so a change in one will change all)
l2[0].append(10)

print(l2)
```

Solution: Don't use * to repeat lists. Make a new one each time.

```python
l = [[] for _ in range(4)]
l[0].append(10)

print(l)
```

#### 5. Class Variables are not Instance Variables.

Think of class variables as static variables in C++.

```python
import datetime

class Person:
    
    population = 0                     # Class Variable (belong to whole class)
    
    def __init__(self, name="<Unknown>", age=20):
        self.name = name                # Instance Variable (belong to calling object)
        self.age = age
        Person.population += 1
    
    def getBirthYear(self):
        return datetime.date.today().year - self.age
    
    def introduce(self):
        print("My name is %s, I was born in %i" % (self.name, self.getBirthYear()))
    
    def __del__(self):
        print("RIP", self.name)
        Person.population -= 1

p = Person("Mohamed", 30)
p.introduce()
print("Population =", Person.population)

p = Person("Ahmed", 20)
p.introduce()
print("Population =", Person.population)

p2 = Person("Samir", 40)
p2.introduce()
print("Population =", Person.population)

del p
del p2
```

#### 6. Class Methods (Static Methods)

An alternative is to just define a function outside the class. But if it's meant to be used with the class, it's better to define it as a class method.

```python
import datetime

class Person:
    
    population = 0
    
    def __init__(self, name="<Unknown>", age=20):
        self.name = name
        self.age = age
        Person.population += 1
    
    def getBirthYear(self):
        return datetime.date.today().year - self.age
    
    def introduce(self):
        print("My name is %s, I was born in %i" % (self.name, self.getBirthYear()))
    
    def __del__(self):
        print("RIP", self.name)
        Person.population -= 1
    
    @staticmethod                   # Decorator for static methods. You can use static variables in them.
    def displayPopulation():
        print("The are %i person(s)" % Person.population)

p = Person("Mohamed", 30)
p.introduce()
Person.displayPopulation()

p = Person("Ahmed", 20)
p.introduce()
Person.displayPopulation()

p2 = Person("Samir", 40)
p2.introduce()
Person.displayPopulation()

del p
del p2
```

#### 7. Using Setters and Getters

This is the formal way of using setters and getters. This way the getter can be also called with the dot operator and backward compatibility is maintained.

```python
class Person:
    
    def __init__(self, name="<Unknown>", age=20):
        self.__name = name                          # Double underscore makes these fully private variables
        self.__age = age if age > 0 else 0 
    
    @property                        # Decorator for getters    
    def name(self):
        return self.__name
    
    @property
    def age(self):
        return self.__age
    
    @age.setter                     # Decorator for setters
    def age(self, age):
        self.__age = age if age > 0 else 0 
    
p1 = Person("Ahmed")
print(p1.name)
print(p1.age)
p1.age = -10
print(p1.age)
p1.age = 20
print(p1.age)
p1.name = "Samir"
```

# XIX. Inheritence

Nothing too fancy, just make sure to call the parent's constructor.

```python
import datetime

class Person:
    def __init__(self, name="<Unknown>", age=20):
        self.name = name
        self.age = age
    
    def getBirthYear(self):
        return datetime.date.today().year - self.age
    
    def introduce(self):
        print("My name is %s, I was born in %i" % (self.name, self.getBirthYear()))

class Student(Person):
    def __init__(self, name="<Unknown>", age=20, school="<Unknown>"):
        super().__init__(name, age)
        self.school = school
    
    def introduce(self):
        print("My name is %s, I was born in %i, I study at %s" % (self.name, self.getBirthYear(), self.school))

class Employee(Person):
    def __init__(self, name="<Unknown>", age=20, company="<Unknown>"):
        super().__init__(name, age)
        self.company = company
    
    def introduce(self):
        print("My name is %s, I was born in %i, I work at %s" % (self.name, self.getBirthYear(), self.company))

p1 = Person("Mohamed", 30)
p1.introduce()

p2 = Student("Ramy", 10, "Future Schools")
p2.introduce()

p3 = Employee("Hoda", 40, "Future Tech Co.")
p3.introduce()

print()
print("is p1 a Person?", isinstance(p1, Person))
print("is p1 a Student?", isinstance(p1, Student))
print("is p1 a Employee?", isinstance(p1, Employee))

print()
print("is p2 a Person?", isinstance(p2, Person))
print("is p2 a Student?", isinstance(p2, Student))
print("is p2 a Employee?", isinstance(p2, Employee))


print()
print("is p3 a Person?", isinstance(p3, Person))
print("is p3 a Student?", isinstance(p3, Student))
print("is p3 a Employee?", isinstance(p3, Employee))
```

# XX. Operator Overloading

#### 1. Overloading Print and +

```python
class Vector2:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return "(%f,%f)" % (self.x, self.y)
    
    def __add__(self, other):
        return Vector2(self.x + other.x, self.y + other.y)
    
v1 = Vector2(1.1, 2.2)
v2 = Vector2(0.9, 0.8)
print(v1, '+', v2, '=', v1+v2)
```

#### 2. Overloading print and *

```python
class Vector2:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return "(%f,%f)" % (self.x, self.y)
    
    def __mul__(self, other):
        if isinstance(other, Vector2):
            return Vector2(self.x * other.x, self.y * other.y)
        else:
            return Vector2(self.x * other, self.y * other)

v1 = Vector2(1.1, 2.2)
v2 = Vector2(0.9, 0.8)
print(v1, '*', v2, '=', v1*v2)
print(v1, '*', 10, '=', v1*10)
print(10, '*', v1, '=', 10*v1)          # Python calls __mul__ of the first operand and an integer doesn't know how it should multiply by your obj.
```

#### 3. Overloading print and * (correctly)

```python
class Vector2:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return "(%f,%f)" % (self.x, self.y)
    
    def __mul__(self, other):
        if isinstance(other, Vector2):
            return Vector2(self.x * other.x, self.y * other.y)
        else:
            return Vector2(self.x * other, self.y * other)
    
    def __rmul__(self, other):
        return self * other                     # We made the string the first operand and called __mul__ (*)

v1 = Vector2(1.1, 2.2)
v2 = Vector2(0.9, 0.8)
print(v1, '*', v2, '=', v1*v2)
print(v1, '*', 10, '=', v1*10)
print(10, '*', v1, '=', 10*v1) # when the scenario above happens, Python calls __rmul__ of the second operand so we just need to implement it!
```
